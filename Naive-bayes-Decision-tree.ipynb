```python
import pandas as pd 

# reading the data
data = pd.read_csv('archive/Breast_Cancer.csv')
```

```python
# showing the data
print('Dataset :',data.shape)
data.info()
data.head()
data.describe()
```

```python
# Remove null columns from data
data.dropna(how='all', axis=1, inplace=True)

# find non-numeric column names
obj_cols = list(data.select_dtypes(include=['object']).columns)

# check if they are useful
for column in obj_cols:
    unique_num = len(data[column].unique())
    total_num = len(data[column])
    uniqueness = (unique_num / total_num) * 100
    # if most of the data is unique, then it's not useful so remove it
    if uniqueness > 90:
        data.drop(column, axis=1, inplace=True)
    # otherwise it's useful, so convert to numeric
    else:
        # map each unique value to it's index
        unique_vals = data[column].unique()
        unique_indexes = range(0, len(unique_vals))
        data[column] = data[column].map(dict(zip(unique_vals, unique_indexes)))
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler

# applying min-max scaler
scaler = MinMaxScaler()
# fitting min-max scaler
data = pd.DataFrame(scaler.fit_transform(data))
```

```python
from sklearn.model_selection import train_test_split

Y = data[15]
X = data.drop(columns=[15])
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
print('X train shape: ', X_train.shape)
print('Y train shape: ', Y_train.shape)
print('X test shape: ', X_test.shape)
print('Y test shape: ', Y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB

# We define the model
gnb = GaussianNB(priors=[0.85, 0.15], var_smoothing=7e-1)

# We train model
gnb.fit(X_train, Y_train)

# We predict target values
Y_predict = gnb.predict(X_test)
```

```python
# Test score
score = gnb.score(X_test, Y_test)
print(score)
```

```python
from sklearn.metrics import classification_report

print(f'{classification_report(Y_test, Y_predict)}')
```

```python
from sklearn.tree import DecisionTreeClassifier

# We define the model
# max_depth=, min_samples_split=, min_samples_leaf=, min_weight_fraction_leaf=, max_features=, min_impurity_decrease=, ccp_alpha=
dt = DecisionTreeClassifier(criterion='entropy', splitter='best', random_state=42, max_leaf_nodes=10, class_weight={0:1.2, 1:1})

# We train model
dt.fit(X_train, Y_train)

# We predict target values
Y_predict2 = dt.predict(X_test)
```

```python
# Test score
score = dt.score(X_test, Y_test)
print(score)
```

```python
print(f'{classification_report(Y_test, Y_predict2)}')
```

```python
from sklearn.ensemble import VotingClassifier

vc = VotingClassifier(estimators=[('naive bayes', gnb), ('decision tree', dt)], voting='soft')
vc.fit(X_train, Y_train)
Y_predict3 = vc.predict(X_test)
```

```python
# Test score
score = vc.score(X_test, Y_test)
print(score)
```

```python
print(f'{classification_report(Y_test, Y_predict3)}')
```
